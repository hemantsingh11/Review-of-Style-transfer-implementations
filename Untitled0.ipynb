```python
import glob
import matplotlib.pyplot as plt
from PIL import Image

content_paths = sorted(glob.glob("./image_files/content/*.jpg"))
style_paths = sorted(glob.glob("./image_files/style/*.jpg"))

# Create a 5x5 grid for the plot
fig, axes = plt.subplots(5, 5, figsize=(20, 20))

# Loop through content and style images
for i, content_path in enumerate(content_paths):
    for j, style_path in enumerate(style_paths):
        # Find the corresponding stylized image
        stylized_path = sorted(glob.glob("./neural_style_transfer/*.jpg"))

        # Open the stylized image and plot it
        stylized_image = Image.open(stylized_path)
        axes[i, j].imshow(stylized_image)
        axes[i, j].axis("off")

# Display the plot
plt.show()
```
