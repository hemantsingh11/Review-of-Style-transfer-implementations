```python
from PIL import Image
import glob
import numpy as np

generated_images = []  # to store all the generated images
for i, content_path in enumerate(sorted(glob.glob("./image_files/content/*.jpg"))):
    first_index = i * 5
    last_index = first_index + 5
    row_images = []  # to store the generated images for each row
    for style_path, stylized_path in zip(sorted(glob.glob("./image_files/style/*.jpg")), sorted(glob.glob("./DFR/*.jpg"))[first_index:last_index]):
        content_image = Image.open(content_path)
        style_image = Image.open(style_path)
        stylized_image = Image.open(stylized_path)
        row_images.extend([content_image, style_image, stylized_image])  # add the images to the row
    row_concatenated = np.concatenate(row_images, axis=1)  # concatenate the images horizontally
    generated_images.append(row_concatenated)  # add the row image to the list of generated images

# concatenate all the generated row images vertically
generated_image = np.concatenate(generated_images, axis=0)

# convert the numpy array to PIL Image object
generated_image = Image.fromarray(generated_image)

# save the generated image as a single JPG file
generated_image.save("generated_images_DFR.jpg")
```
